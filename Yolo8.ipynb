```python
model_path = '/project/ece601/ec601_lifeguard/runs/detect/train7/weights/best.pt'
image_path = '/project/ece601/ec601_lifeguard/Drowning.png'

from ultralytics import YOLO
import cv2
from matplotlib import pyplot as plt
import torch

# Âä†ËΩΩÊ®°ÂûãÂíåÊùÉÈáç
model = YOLO('/project/ece601/ec601_lifeguard/runs/detect/train14/weights/best.pt')

# ËØªÂèñÂõæÂÉè
image_path = '/project/ece601/ec601_lifeguard/3.jpg'
image = cv2.imread(image_path)

# ËøõË°åÊé®ÁêÜ
results = model(image)

# Ëé∑ÂèñËæπÁïåÊ°Ü„ÄÅÁΩÆ‰ø°Â∫¶ÂíåÁ±ªÂà´‰ø°ÊÅØ
boxes = results[0].boxes.xyxy  # ËæπÁïåÊ°ÜÂùêÊ†á
confidences = results[0].boxes.conf  # ÁΩÆ‰ø°Â∫¶
classes = results[0].boxes.cls  # Á±ªÂà´

# Â∞ÜTensorËΩ¨Êç¢‰∏∫NumPyÊï∞ÁªÑÔºàÂ¶ÇÊûúÈúÄË¶ÅÔºâ
if isinstance(boxes, torch.Tensor):
    boxes = boxes.cpu().numpy()
if isinstance(confidences, torch.Tensor):
    confidences = confidences.cpu().numpy()
if isinstance(classes, torch.Tensor):
    classes = classes.cpu().numpy()

# Âú®ÂõæÁâá‰∏äÁªòÂà∂ËæπÁïåÊ°ÜÂíåÁõ∏ÂÖ≥‰ø°ÊÅØ
for i, box in enumerate(boxes):
    x1, y1, x2, y2 = box
    conf = confidences[i]
    cls = classes[i]
    cv2.rectangle(image, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
    label = f'Class: {int(cls)}, Conf: {conf:.2f}'
    cv2.putText(image, label, (int(x1), int(y1-10)), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)

# Â∞ÜBGRËΩ¨Êç¢‰∏∫RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# ‰ΩøÁî®matplotlibÊòæÁ§∫ÂõæÂÉè
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
from ultralytics import YOLO
import cv2
from matplotlib import pyplot as plt
import torch
from transformers import BertModel, BertTokenizer

# Initial BERT
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model_bert = BertModel.from_pretrained('bert-base-uncased')

# Load YOLO
yolo_model_path = '/project/ece601/ec601_lifeguard/runs/detect/train14/weights/best.pt'
yolo_model = YOLO(yolo_model_path)

image_path = '/project/ece601/ec601_lifeguard/3.jpg'
image = cv2.imread(image_path)

results = yolo_model(image)

for i, detection in enumerate(results[0].boxes.xyxy):
    x1, y1, x2, y2 = detection[:4]
    conf = results[0].boxes.conf[i].clone()
    cls = results[0].boxes.cls[i]
    label = yolo_model.names[int(cls)]

    if label == 'drowning':
        conf += 0.15
        conf = min(conf, 1.0)

    inputs = tokenizer(label, return_tensors="pt")
    outputs = model_bert(**inputs)
    
    # BERT
    cv2.rectangle(image, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
    cv2.putText(image, f"{label}: {conf:.2f}", (int(x1), int(y1 - 10)), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)

# #Change BGR into RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```
