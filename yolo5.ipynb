```python
# 假设的图像尺寸
image_width, image_height = 1280, 720

# 示例标注数据
label_data = [
    # ... 其他标注数据
    [444,105,447,475],[973,292,343,680],
[417,97,502,585],[991,253,359,670],
[429,122,435,575],[991,397,337,595],
[437,110,489,505],[1016,410,327,587],
[407,110,524,535],[1061,472,307,525],
[437,102,424,575],[1056,452,292,490],
[407,122,524,448],[1063,470,273,402],
[407,110,519,547],[1053,430,320,402],
[467,115,349,540],[1053,482,307,348],
[407,92,512,553],[1011,455,337,317],
[414,110,505,595],[991,460,364,335],
[449,130,370,472],[986,470,350,310],
[407,97,517,568],[1006,482,312,295],
[412,102,494,628],[1006,442,335,340],
[432,120,472,470],[981,467,372,388],
[407,92,519,593],[1038,435,280,342],
[437,90,432,620],[1026,432,354,373],
[424,105,490,497],[1006,420,404,452],
[412,110,507,555],[968,380,460,417],
[474,110,345,575],[966,340,449,477],
[419,102,520,495],[993,372,437,475],
[404,105,525,597],[991,347,335,643],
[474,110,392,547],[978,172,390,673],
[412,115,519,490],[1031,395,324,595],
[407,110,522,580],[1036,467,319,528],
[467,110,352,537],[1063,405,305,547],
[412,110,517,555],[1066,447,289,470],
[404,102,515,643],[1078,410,265,505]
]

def convert_to_yolov5(label, img_width, img_height, class_id):
    x_min, y_min, width, height = label
    x_center = (x_min + width / 2) / img_width
    y_center = (y_min + height / 2) / img_height
    width /= img_width
    height /= img_height
    return [class_id, x_center, y_center, width, height]

# 转换所有标注
converted_labels = []
for i in range(0, len(label_data), 2):
    label1 = convert_to_yolov5(label_data[i], image_width, image_height, 0)  # 类别 0
    label2 = convert_to_yolov5(label_data[i+1], image_width, image_height, 1)  # 类别 1
    converted_labels.append([label1, label2])  # 每个图像有两个标注

# 输出转换后的标注
for labels in converted_labels:
    for label in labels:
        print(label)
```

```python
import os


# 存储 .txt 文件的目录
output_dir = '/project/ece601/ec601_lifeguard/image'  # 更改为您的目录路径

# 确保输出目录存在
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# 为每个图像创建一个对应的 .txt 文件
for i, labels in enumerate(converted_labels, 1):
    txt_filename = os.path.join(output_dir, f'img{i}.txt')
    with open(txt_filename, 'w') as file:
        for label in labels:
            label_str = ' '.join(map(str, label))
            file.write(label_str + '\n')

    print(f"File created: {txt_filename}") 

```
